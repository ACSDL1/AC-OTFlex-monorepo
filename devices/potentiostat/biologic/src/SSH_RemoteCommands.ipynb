# SSH Remote Command Execution
Execute various system commands on a remote machine (sdl1_@192.168.0.102) using Paramiko SSH client.

## Import Required Libraries
Import Paramiko and other necessary libraries for SSH connection and output handling.

```python
import paramiko
import warnings

# Suppress SSH warnings
warnings.filterwarnings('ignore')
```

## Establish SSH Connection
Create an SSH client using Paramiko and connect to sdl1_@192.168.0.102 with authentication.

```python
# SSH connection parameters
hostname = "192.168.0.102"
username = "sdl1_"
port = 22

# Create SSH client
ssh_client = paramiko.SSHClient()
ssh_client.set_missing_host_key_policy(paramiko.AutoAddPolicy())

try:
    # Connect to the remote machine
    # Note: Update password or use SSH key authentication as needed
    ssh_client.connect(
        hostname=hostname,
        port=port,
        username=username,
        password="your_password_here"  # Replace with actual password or use key-based auth
    )
    print(f"✓ Connected to {username}@{hostname}")
except Exception as e:
    print(f"✗ Connection failed: {e}")
```

## Execute Fastfetch Command
Execute the fastfetch command on the remote machine to display system information.

```python
# Execute fastfetch command
try:
    stdin, stdout, stderr = ssh_client.exec_command("fastfetch")
    output = stdout.read().decode()
    error = stderr.read().decode()
    
    print("=== FASTFETCH OUTPUT ===")
    print(output)
    
    if error:
        print("=== ERRORS ===")
        print(error)
except Exception as e:
    print(f"✗ Failed to execute fastfetch: {e}")
```

## List Installed Python Libraries with Pip
Execute pip list command to display all installed Python packages and their versions.

```python
# Execute pip list command
try:
    stdin, stdout, stderr = ssh_client.exec_command("pip list")
    output = stdout.read().decode()
    error = stderr.read().decode()
    
    print("=== PIP LIST OUTPUT ===")
    print(output)
    
    if error:
        print("=== ERRORS ===")
        print(error)
except Exception as e:
    print(f"✗ Failed to execute pip list: {e}")
```

## List Conda Environments
Execute conda env list command to display all available Conda environments on the remote machine.

```python
# Execute conda env list command
try:
    stdin, stdout, stderr = ssh_client.exec_command("conda env list")
    output = stdout.read().decode()
    error = stderr.read().decode()
    
    print("=== CONDA ENV LIST OUTPUT ===")
    print(output)
    
    if error:
        print("=== ERRORS ===")
        print(error)
except Exception as e:
    print(f"✗ Failed to execute conda env list: {e}")
finally:
    # Close the SSH connection
    ssh_client.close()
    print("\n✓ SSH connection closed")
```
