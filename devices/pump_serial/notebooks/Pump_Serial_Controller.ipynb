# Pump Serial Controller

Control ESP32-based quad relay pump driver via serial connection.

## Hardware Setup
- **Pump 1**: negative on relay 1, positive on relay 2
- **Pump 2**: negative on relay 3, positive on relay 4

## Relay Wiring
- All NO (normally open) contacts → +12V
- All NC (normally closed) contacts → -12V (connected together)

## Pump Control Logic
- Pump ON: Opposite relay states (one ON, one OFF) → 24V potential difference
- Pump OFF: Same relay states (both ON or both OFF) → 0V

## Communication
- Serial protocol at 115200 baud
- Commands sent as text with newline terminator

```python
import sys
import time
import os
from pathlib import Path

# System-agnostic path resolution
# Walks up directory tree to find pump_serial/src directory
def find_pump_serial_root():
    """Find pump_serial root by looking for src/pump_serial.py"""
    current = Path.cwd()
    for parent in [current, current.parent, current.parent.parent, current.parent.parent.parent]:
        if (parent / 'src' / 'pump_serial.py').exists():
            return parent
    return current

pump_serial_root = find_pump_serial_root()
sys.path.insert(0, str(pump_serial_root))

from src.pump_serial import PumpSerial
```

## Initialize Connection

Connect to the pump controller on the specified serial port.

```python
# Initialize pump controller
SERIAL_PORT = "/dev/ttyUSB0"  # Change this to your serial port

pump = PumpSerial(SERIAL_PORT)
print("Pump controller connected!")
```

## Get Status

Check current pump states.

```python
pump.status()
```

## Control Pumps

### Turn on Pump 1 for 3 seconds (3000 ms)

```python
print("Turning pump 1 on for 3000 ms...")
pump.on(1, 3000)
print("Waiting for pump to stop...")
time.sleep(4)
print("Done!")
```

### Turn on Pump 2 for 2 seconds (2000 ms)

```python
print("Turning pump 2 on for 2000 ms...")
pump.on(2, 2000)
print("Waiting for pump to stop...")
time.sleep(3)
print("Done!")
```

## Sequential Pump Control

Example of controlling multiple pumps in sequence.

```python
# Run pump 1 for 1 second, then pump 2 for 1 second
print("Sequence test: Pump 1 (1s) -> Pump 2 (1s)")

print("Starting pump 1...")
pump.on(1, 1000)
time.sleep(1.5)

print("Starting pump 2...")
pump.on(2, 1000)
time.sleep(1.5)

print("Sequence complete!")
```

## Close Connection

Cleanly close the serial connection when done.

```python
pump.close()
print("Connection closed.")
```
