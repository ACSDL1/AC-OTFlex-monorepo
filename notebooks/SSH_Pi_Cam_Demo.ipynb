# Raspberry Pi Camera Demo
Minimal notebook that uses the `src.adapters.pi_cam` adapter functions to capture, optionally rotate, and display an image.

```python
from pathlib import Path
import sys

proj_root = Path.cwd()
while proj_root.name != "AC-OTFlex-monorepo" and proj_root.parent != proj_root:
    proj_root = proj_root.parent

if str(proj_root) not in sys.path:
    sys.path.insert(0, str(proj_root))

from src.adapters.pi_cam import capture_pi_image_via_ssh, rotate_image_if_needed

PI_HOST = "192.168.0.101"
PI_USER = "sdl1"
PI_PASS = "1144"
ROTATE_IMAGE = True
```

```python
image_path = capture_pi_image_via_ssh(
    PI_HOST,
    PI_USER,
    PI_PASS,
    output_dir=proj_root / "data" / "out" / "images",
)

rotate_image_if_needed(image_path, rotate=ROTATE_IMAGE, degrees=180)

print(f"Saved image: {image_path}")
```
